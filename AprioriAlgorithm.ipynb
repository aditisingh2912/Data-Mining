```python
from itertools import combinations
transactions =
{
    "T100": {"I1", "I2", "I5"},
    "T200": {"I2", "I4"},
    "T300": {"I2", "I3"},
    "T400": {"I1", "I2", "I4"},
    "T500": {"I1", "I3"},
    "T600": {"I2", "I3"},
    "T700": {"I1", "I3"},
    "T800": {"I1", "I2", "I3", "I5"},
    "T900": {"I1", "I2", "I3"}
}
min_support_count = 2
min_confidence = 0.7

def get_support(itemset, transactions):
    return sum(1 for transaction in transactions.values() if itemset.issubset(transaction))


unique_items = sorted(set(item for transaction in transactions.values() for item in transaction))
item_counts = {item: get_support({item}, transactions) for item in unique_items}
frequent_items = {item for item, count in item_counts.items() if count >= min_support_count}

frequent_itemsets = []
current_itemsets = [{item} for item in frequent_items]

while current_itemsets:
    itemset_supports = {frozenset(itemset): get_support(set(itemset), transactions) for itemset in current_itemsets}
    current_frequent_itemsets = {itemset: support for itemset, support in itemset_supports.items() if support >= min_support_count}

    if not current_frequent_itemsets:
        break

    frequent_itemsets.append(current_frequent_itemsets)

    current_itemsets = [set(a).union(set(b)) for a, b in combinations(current_frequent_itemsets.keys(), 2) if len(a.union(b)) == len(a) + 1]

final_frequent_itemsets = {itemset: support for level in frequent_itemsets for itemset, support in level.items()}

association_rules_list = []
for itemset, support in final_frequent_itemsets.items():
    if len(itemset) > 1:
        for lhs_size in range(1, len(itemset)):
            for lhs in combinations(itemset, lhs_size):
                lhs = set(lhs)
                rhs = itemset - lhs
                lhs_support = get_support(lhs, transactions)

                confidence = support / lhs_support if lhs_support else 0
                if confidence >= min_confidence:
                    association_rules_list.append((lhs, rhs, confidence))

print("Frequent Itemsets:")
for itemset, support in final_frequent_itemsets.items():
    print(f"{set(itemset)}: {support}")

print("\nStrong Association Rules:")
for lhs, rhs, confidence in association_rules_list:
    print(f"{lhs} → {rhs} (Confidence: {confidence*100:.2f}%)")
```

```python
from itertools import combinations
transactions = [
    {'M', 'O', 'N', 'K', 'E', 'Y'},
    {'D', 'O', 'N', 'K', 'E', 'Y'},
    {'M', 'A', 'K', 'E'},
    {'M', 'U', 'C', 'K', 'Y'},
    {'C', 'O', 'O', 'K', 'I', 'E'}
]

min_support_count = 3
min_confidence = 0.8

item_counts = {}
for transaction in transactions:
    for item in transaction:
        item_counts[item] = item_counts.get(item, 0) + 1

frequent_items = {item for item, count in item_counts.items() if count >= min_support_count}


frequent_itemsets = {}
for itemset in combinations(frequent_items, 2):
    count = sum(1 for transaction in transactions if set(itemset).issubset(transaction))
    if count >= min_support_count:
        frequent_itemsets[itemset] = count


for itemset in combinations(frequent_items, 3):
    count = sum(1 for transaction in transactions if set(itemset).issubset(transaction))
    if count >= min_support_count:
        frequent_itemsets[itemset] = count

strong_rules = []
for itemset, sup_count in frequent_itemsets.items():
    for i in range(1, len(itemset)):
        for lhs in combinations(itemset, i):
            rhs = tuple(set(itemset) - set(lhs))
            lhs_count = item_counts[lhs[0]] if len(lhs) == 1 else frequent_itemsets.get(lhs, 0)
            if lhs_count > 0:
                confidence = sup_count / lhs_count
                if confidence >= min_confidence:
                    strong_rules.append((lhs, rhs, confidence))


print("Frequent Itemsets:")
for itemset, count in frequent_itemsets.items():
    print(f"{itemset}: {count}")

print("\nStrong Association Rules:")
for lhs, rhs, confidence in strong_rules:
    print(f"{lhs} → {rhs} (Confidence: {confidence:.2%})")
```
